````python
def makePretty(p, color):
    p.SetStats(0)
    p.SetMarkerColor(color)
    p.SetLineColor(color)
    p.SetFillColor(color)
    p.SetFillStyle(3003)
    p.SetMarkerStyle(1)
    p.SetMarkerSize(0.2)
    p.SetTitle("")

def GetKeyNames( self, dir = "" ):
        self.cd(dir)
        return [key.GetName() for key in gDirectory.GetListOfKeys()]
````

````python
TFile.GetKeyNames = GetKeyNames
keyList = file1.GetKeyNames()
print(keyList)
````

````python
h_mc_quark_pp = file1.Get("h_mc_quark_pp")
h_mc_quark_p  = file1.Get("h_mc_quark_p")

makePretty(h_mc_quark_pp,kBlack)
makePretty(h_mc_quark_p,kBlack)
````
