````python
def makePretty(p, color):
    p.SetStats(0)
    p.SetMarkerColor(color)
    p.SetLineColor(color)
    p.SetFillColor(color)
    p.SetFillStyle(3003)
    p.SetMarkerStyle(1)
    p.SetMarkerSize(0.2)
    p.SetTitle("")

def GetKeyNames( self, dir = "" ):
        self.cd(dir)
        return [key.GetName() for key in gDirectory.GetListOfKeys()]
    
def Normal(h1):
    h1.Scale(1.0/h1.GetEntries())
    
def DivBin(h1):
    for ibin in range(h1.GetNbinsX()):
        binCont = h1.GetBinContent(ibin)
        binWid  = h1.GetXaxis().GetBinWidth(ibin)
        binContnew = binCont/binWid
        h1.SetBinContent(ibin,binContnew)
        
````

````python
TFile.GetKeyNames = GetKeyNames
keyList = file1.GetKeyNames()
print(keyList)
````

````python
h_pfo_jet_angdiff    = file1.Get("h_pfo_jet_angdiff")

h_pfo_jet_charge_u     = file1.Get("h_pfo_jet_charge_u")
h_pfo_jet_charge_ubar = file1.Get("h_pfo_jet_charge_ubar")

h_pfo_jet_charge_d     = file1.Get("h_pfo_jet_charge_d")
h_pfo_jet_charge_dbar = file1.Get("h_pfo_jet_charge_dbar")

h_pfo_jet_charge_s     = file1.Get("h_pfo_jet_charge_s")
h_pfo_jet_charge_sbar = file1.Get("h_pfo_jet_charge_sbar")

h_pfo_q_match_charge    = file1.Get("h_pfo_q_match_charge")
h_pfo_qbar_match_charge = file1.Get("h_pfo_qbar_match_charge")

makePretty(h_pfo_jet_angdiff,kBlack)

makePretty(h_pfo_jet_charge_u,kRed)
makePretty(h_pfo_jet_charge_ubar,kBlue)

makePretty(h_pfo_jet_charge_d,kRed)
makePretty(h_pfo_jet_charge_dbar,kBlue)

makePretty(h_pfo_jet_charge_s,kRed)
makePretty(h_pfo_jet_charge_sbar,kBlue)

makePretty(h_pfo_q_match_charge,kRed)
makePretty(h_pfo_qbar_match_charge,kBlue)
````

````python
# Normal(h_pfo_jet_charge)
# Normal(h_pfo_jet_charge_bar)
DivBin(h_pfo_jet_charge_u)
DivBin(h_pfo_jet_charge_ubar)

DivBin(h_pfo_jet_charge_d)
DivBin(h_pfo_jet_charge_dbar)

DivBin(h_pfo_jet_charge_s)
DivBin(h_pfo_jet_charge_sbar)
````

````python
def JetChargeHist(h1,h2,quark,Lcvs):
    canvas = TCanvas(h1.GetName(),h1.GetName(),500,500)
    pad = canvas.cd(0)
#     iobj.SetTitle(iobj.GetName())
    h1.Draw("h")
    h2.Draw("hsame")
    h1.SetTitle(quark)
    h2.SetTitle("#bar{"+quark+"}")    
    pad.BuildLegend(0.75, 0.9, 0.9, 0.8)
    Lcvs.append(canvas)
````

````python
Lcvs = []
JetChargeHist(h_pfo_jet_charge_u,h_pfo_jet_charge_ubar,"u",Lcvs)
JetChargeHist(h_pfo_jet_charge_d,h_pfo_jet_charge_dbar,"d",Lcvs)
JetChargeHist(h_pfo_jet_charge_s,h_pfo_jet_charge_sbar,"s",Lcvs)
for icvs in Lcvs:
    icvs.Draw()
````
