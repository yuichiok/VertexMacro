```python
def makePretty(p, color):
    p.SetStats(0)
    p.SetLineColor(color)
    p.SetTitle("")
    
def makePretty2D(p, color):
    p.SetStats(0)
    p.SetLineColor(color)
    p.SetMarkerColor(color)
    p.SetTitle("")

def Normal(h1):
    h1.Scale(1.0/h1.GetEntries())
```
