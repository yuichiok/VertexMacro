```python
gStyle.SetOptFit(0)
gStyle.SetOptStat(0)  
gStyle.SetOptTitle(1)
gStyle.SetTitleBorderSize(0)
gStyle.SetTitleStyle(0)
gStyle.SetMarkerSize(1)
gStyle.SetTitleX(0.2) 
gStyle.SetTitleY(0.9)
gStyle.SetPadLeftMargin(0.2)
```

```python
def makePretty(p, color):
    p.SetStats(0)
    p.SetLineColor(color)
    p.SetTitle("")

def Normal(h1):
    h1.Scale(1.0/h1.GetEntries())
```

```python
file1 = TFile.Open("/home/ilc/yokugawa/macros/QQbar250/analysis/ssbar/Hit/DQ_250GeV_test.root")
```

```python
h_pfo_k_2_10_cos  = file1.Get("h_pfo_k_2_10_cos")
h_pfo_k_10_30_cos = file1.Get("h_pfo_k_10_30_cos")
h_pfo_k_30_cos    = file1.Get("h_pfo_k_30_cos")

h_pfo_LeadK_2_10_cos  = file1.Get("h_pfo_LeadK_2_10_cos")
h_pfo_LeadK_10_30_cos = file1.Get("h_pfo_LeadK_10_30_cos")
h_pfo_LeadK_30_cos    = file1.Get("h_pfo_LeadK_30_cos")

h_list_k = []
h_list_leadK = []

h_list_k.append(h_pfo_k_2_10_cos)
h_list_k.append(h_pfo_k_10_30_cos)
h_list_k.append(h_pfo_k_30_cos)

h_list_leadK.append(h_pfo_LeadK_2_10_cos)
h_list_leadK.append(h_pfo_LeadK_10_30_cos)
h_list_leadK.append(h_pfo_LeadK_30_cos)

for ih in h_list_k:
    makePretty(ih,kBlue)
    Normal(ih)
    
for ih in h_list_leadK:
    makePretty(ih,kRed)
    Normal(ih)
```

```python
lc = []

for i in range(3):
    c_name = "c" + str(i)
    c = TCanvas(c_name,c_name,550,500)
    h_list_leadK[i].Draw("h")
    h_list_k[i].Draw("hsame")
    lc.append(c)
    
for i in lc:
    i.Draw()
    
```
